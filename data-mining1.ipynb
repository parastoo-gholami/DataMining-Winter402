```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data set-1.csv')
df
```

```python
def statistic(column):
    value= column.value_counts()
    #print(value)
    type_col=len(value)
    most_word=value.idxmax()
    most_count=value.max()
    least_word=value.idxmin()
    least_count=value.min()
    missing=column.isna().sum()
    return type_col,missing, most_word,most_count,least_word,least_count
#tekrar- most- count-least-count
statics=df.apply(statistic)
statics.iloc[0]= statics.iloc[0].apply(lambda value: 'biominal' if value ==2 else 'polynominal' if value>2 else value)
statics
```

```python
#change 0
name=df.columns[1]
value= df[name].value_counts()
value.plot(kind='bar', xlabel='words', ylabel='frequency', title=f'frequency of words column {name}')
```

```python
#replace
df['Twitter'].replace('99', pd.NA , inplace=True)
df
```

```python
#hazf
df.dropna(subset=['Online_Shopping'], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
#fill with max
df['Read_News'].fillna(df['Read_News'].value_counts().idxmax(), inplace=True)
df
```

```python
#fill with mean
df['Years_on_Internet'].fillna(df['Years_on_Internet'].mean(), inplace=True)
df
```

```python
#fill with middle
df['Hours_Per_Day'].fillna(df['Hours_Per_Day'].quantile(0.5), inplace=True)
df 
```
