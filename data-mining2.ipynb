## Imports

```python
import pandas as pd
import numpy as np
from plotly import graph_objects as go
```

## Read Data

```python
house_warm = pd.read_csv(r"C:\Users\User\Desktop\data\data set-2.csv")
house_warm
```

## Calculate Correlation

```python
correlation = pd.DataFrame(index=range(6),columns=['Name','Insulation','Temperature','Heating_Oil','Num_Occupants','Avg_Age','Home_Size'])
name = ['Insulation','Temperature','Heating_Oil','Num_Occupants','Avg_Age','Home_Size']

for i in range(len(correlation)):
    correlation['Name'][i] = name[i]
    correlation['Insulation'][i] = house_warm['Insulation'].corr(house_warm[name[i]])
    correlation['Temperature'][i] = house_warm['Temperature'].corr(house_warm[name[i]])
    correlation['Heating_Oil'][i] = house_warm['Heating_Oil'].corr(house_warm[name[i]])
    correlation['Num_Occupants'][i] = house_warm['Num_Occupants'].corr(house_warm[name[i]])
    correlation['Avg_Age'][i] = house_warm['Avg_Age'].corr(house_warm[name[i]])
    correlation['Home_Size'][i] = house_warm['Home_Size'].corr(house_warm[name[i]])

correlation
```
